```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import SpectralClustering
```

```python
# Load the dataset
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, 3:5].values
```

```python

# Determine the optimal number of clusters (optional)
# Spectral clustering doesn't have a direct equivalent to the Elbow Method
# You can use techniques like silhouette analysis or domain knowledge to choose the number of clusters

# Apply Spectral Clustering
n_clusters = 5  # Replace with your chosen number of clusters
spectral = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors')  # Adjust affinity if needed
y_spectral = spectral.fit_predict(X)
```

```python
# Add cluster labels to the dataset
supervised = dataset
supervised['Cluster_group'] = y_spectral
```

```python
# Save the clustered dataset
supervised.to_csv("ML_cluster_spectral.csv", index=False)
```

```python
# Visualize the clusters
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5],
                   fit_reg=False, legend=True, legend_out=True)
```
