```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import AffinityPropagation
```

```python
# Load the dataset
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, 3:5].values
```

```python
# Create the Affinity Propagation model
ap = AffinityPropagation(damping=0.5, preference=-50)  # Adjust damping and preference as needed
```

```python
# Fit the model to the data
ap.fit(X)
```

```python
# Get the cluster labels
cluster_labels = ap.labels_
```

```python
# Add cluster labels to the original dataset
supervised = dataset
supervised['Cluster_group'] = cluster_labels
```

```python
# Save the clustered dataset
supervised.to_csv("ML_cluster_AP.csv", index=False)
```

```python
# Plot the clusters using seaborn (similar to your K-means code)
import seaborn as sns
facet = sns.lmplot(
   data=supervised,
   x=supervised.columns[3],
   y=supervised.columns[4],
   hue=supervised.columns[5],
   fit_reg=False,
   legend=True,
   legend_out=True
)
```
